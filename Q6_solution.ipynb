```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(theta1, theta2, R, A, w):
    return (2 * np.pi * A * (1 - np.cos(theta1 - theta2)))/(R * w)

R = 1000 # Resistor value, unit is ohm
A = 1    # Voltage amplitude, unit is V
w = np.pi    # Radial frequency, unit is radians per second

theta1 = np.linspace(-np.pi, np.pi, 100) # Sweep theta1 from -pi to pi
theta2 = np.linspace(-np.pi, np.pi, 100) # Sweep theta2 from -pi to pi
Theta1, Theta2 = np.meshgrid(theta1, theta2) # Convert to 2D domain

Energy = f(Theta1, Theta2, R, A, w) # Calculate energy

# Start to plot
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(Theta1, Theta2, Energy, 80, cmap='ocean')
ax.set_xlabel('Theta_1')
ax.set_ylabel('Theta_2')
ax.set_zlabel('E');
```

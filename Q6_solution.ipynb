```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(theta1, theta2, R, A, w):
    return (2 * np.pi * A * (1 - np.cos(theta1 - theta2)))/(R * w)

R = 1000
A = 1
w = 1

theta1 = np.linspace(-np.pi, np.pi, 100)
theta2 = np.linspace(-np.pi, np.pi, 100)

Theta1, Theta2 = np.meshgrid(theta1, theta2)

Energy = f(Theta1, Theta2, R, A, w)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(Theta1, Theta2, Energy, 100, cmap='ocean')
ax.set_xlabel('Theta_1')
ax.set_ylabel('Theta_2')
ax.set_zlabel('E');
```
